# Data visualization:

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('data.csv')
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
g = sns.catplot(data=df, x="Age", y="AttackMethod", hue="Maritalstatus", kind="swarm", orient="h")
g.fig.set_figwidth(25)
g.fig.set_figheight(4)
```

```python
sns.countplot(y="Harm", data=df, palette="husl")
```

```python
df['Age'].hist(figsize=(5,5), color='brown', alpha=1)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age and its Count')
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'AttackMethod')
```

```python
piechart(df, 'Gender')
```

```python
g = sns.factorplot("Attack", data=df, aspect=5.0, kind='count', hue='Age')
g.set_ylabels('Number of persons')
g.fig.set_figheight(10)
```

```python
sns.scatterplot(x="Income", y="Harm", data=df)
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Crime', 'Gender', 'Age', 'Income', 'Job', 'Maritalstatus', 'Education', 'Harm', 'Attack', 'AttackMethod']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
fig, ax = plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot = True, ax=ax)
plt.show()
```
